```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
```

```python
df = sns.load_dataset("titanic")    # Bu seaborn'da hazır bulunan titanic datasetini hazır olarak yükleme yöntemidir.
```

```python
df
```

```python
df[["sex","class","embark_town"]].isnull().sum()
```

```python
df = df.dropna(subset = 'embark_town')    # embark_town değişkeninde iki tane NaN değer vardı bunları silmek için bunu yaptım.
```

------------------
## LABEL ENCODER

```python
label_encoder = LabelEncoder()
```

```python
df_label = df.copy()
```

```python
df_label
```

```python
df_label["sex"] = label_encoder.fit_transform(df_label["sex"])    # df_label dataframe'i içerisine df_label["sex"] kolonunun encode edilmiş halini koyuyoruz.
```

```python
df_label
```

-----
## ORDINAL ENCODER

```python
df_ordinal = df.copy()
```

```python
class_order = ["Third","Second","First"]
```

```python
ordinal_encoder = OrdinalEncoder(categories=[class_order])
```

```python
df_ordinal["class"] = ordinal_encoder.fit_transform(df_ordinal[["class"]])
```

```python
df_ordinal
```

Burada ise class yerine ordinal encoder yapmış olduk. 1, 2, 3 gibi değerler vermiş olduk.
