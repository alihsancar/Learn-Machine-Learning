```python
df = df[(df.weight_kg > 35) & (df.weight_kg < 120)]
df = df[(df['body fat_%'] > 5) & (df['body fat_%'] < 55)]
df = df[(df['diastolic'] > 20) & (df['diastolic'] < 140)]
df = df[(df['systolic'] > 50)]
df = df[(df['gripForce'] > 2)]
df = df[(df['sit and bend forward_cm'] < 150)]
df = df[df['broad jump_cm'] > 25]
```

```python
df.describe()
```

# train - test split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('class', axis = 1)
y = df['class']
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=15)
```

# Encoding

```python
X_train['gender'].value_counts()
```

```python
X_train = pd.get_dummies(X_train, columns=['gender'], dtype=int)
```

```python
X_train.head()
```

```python
X_test = pd.get_dummies(X_test, columns=['gender'], dtype=int)
```

```python
X_test.head()
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
logistic = LogisticRegression()
logistic.fit(X_train_scaled, y_train)
y_pred = logistic.predict(X_test_scaled)
print("Accuracy Score :", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Confusion Matrix :\n", confusion_matrix(y_test, y_pred))
```

# OnevsOne - OnevsRest

```python
from sklearn.multiclass import OneVsOneClassifier, OneVsRestClassifier
```

```python
onevsrest = OneVsRestClassifier(LogisticRegression())
onevsrest.fit(X_train_scaled, y_train)
y_pred = onevsrest.predict(X_test_scaled)
print("Accuracy Score :", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Confusion Matrix :\n", confusion_matrix(y_test, y_pred))
```

```python
onevsone = OneVsOneClassifier(LogisticRegression())
onevsone.fit(X_train_scaled, y_train)
y_pred = onevsone.predict(X_test_scaled)
print("Accuracy Score :", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Confusion Matrix :\n", confusion_matrix(y_test, y_pred))
```

# SVM Classifaction

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
svc.fit(X_train_scaled, y_train)
y_pred = svc.predict(X_test_scaled)
print("Accuracy Score :", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Confusion Matrix :\n", confusion_matrix(y_test, y_pred))
```

## Hyperparameter Tuning for SVC

```python
from sklearn.model_selection import GridSearchCV ,StratifiedKFold
```

```python
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100, 1000],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto', 0.01, 0.1, 1]
}
cv = StratifiedKFold()
```

```python
grid = GridSearchCV(estimator=SVC(), param_grid=param_grid, cv=cv, n_jobs=-1, refit=True,verbose=3)
```

```python
grid.fit(X_train_scaled, y_train)
y_pred = grid.predict(X_test_scaled)
print("Accuracy Score :", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Confusion Matrix :\n", confusion_matrix(y_test, y_pred))
```
