# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=15)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
def correlation_for_dropping(df, threshold):
    columns_to_drop = set()
    corr = df.corr()

    for i in range(len(corr.columns)):
        for j in range(i):
            if abs(corr.iloc[i,j]) > threshold:
                columns_to_drop.add(corr.columns[i])

    return columns_to_drop
```

```python
columns_to_drop = correlation_for_dropping(X_train, 0.85)
```

```python
columns_to_drop
```

Boş gelmesi demek train veri seti içinde yer alan değişkenlerin kendi aralarındaki korelasyonları çok yüksek değil yani silmeye gerek yok demek.

# Scaling İşlemi

```python
from sklearn.preprocessing import StandardScaler, RobustScaler
```

```python
scaler = StandardScaler()
X_train_Standart = scaler.fit_transform(X_train)
X_test_Standart = scaler.transform(X_test)
```

```python
robust = RobustScaler()
X_train_Robust = robust.fit_transform(X_train)
X_test_Robust = robust.transform(X_test)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
linear = LinearRegression()
```

```python
# Standart Scaling ile
linear.fit(X_train_Standart, y_train)
```

```python
y_pred = linear.predict(X_test_Standart)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Absolute Error : " , mae)
print("Mean Squared Error : " , mse)
print("R2 score : " , r2)
```

```python
# Robust Scaling ile
robustLinear = LinearRegression()
robustLinear.fit(X_train_Robust, y_train)
```

```python
y_pred = robustLinear.predict(X_test_Robust)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Absolute Error : " , mae)
print("Mean Squared Error : " , mse)
print("R2 score : " , r2)
```

```python
plt.scatter(y_test, y_pred)
plt.show()
```
